```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
from sklearn import cluster
```

```python
orig_img = mpimg.imread('color-gamut-srgb-large.jpg')
```

```python
base = orig_img.shape[0];
height = orig_img.shape[1];
channels = orig_img.shape[2]
```

```python
print('image size: ' + str(base) + 'x' + str(height))
print('number of channels (rgb): ' + str(channels))
```

```python
orig = orig_img.reshape(base * height, channels)
```

```python
kmcluster = cluster.KMeans(n_clusters=3)
kmcluster.fit(orig)
cluster_centers = kmcluster.cluster_centers_
cluster_labels = kmcluster.labels_
```

```python
plt.imshow(cluster_centers[cluster_labels].reshape(base, height, channels).astype(np.uint8))
```
